```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("diabetes.csv")

df.head()
```

```python
print(df.shape)

df.info()

df.describe()
```

```python
X = df.drop("Outcome", axis=1)

y = df["Outcome"]

print(X.head())
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy without PCA/LDA :",
      accuracy_score(y_test, y_pred))
```

```python
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

print("Original Shape :", X_scaled.shape)
print("Reduced Shape :", X_pca.shape)
```

```python
print("Explained Variance Ratio:")

print(pca.explained_variance_ratio_)

print("Total Variance Explained:")

print(sum(pca.explained_variance_ratio_))
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(
    X_pca,
    y,
    test_size=0.2,
    random_state=42
)

model_pca = LogisticRegression()

model_pca.fit(X_train_pca, y_train)

pred_pca = model_pca.predict(X_test_pca)

print("Accuracy after PCA :")

print(accuracy_score(y_test, pred_pca))
```

```python
lda = LinearDiscriminantAnalysis(n_components=1)

X_lda = lda.fit_transform(X_scaled, y)

print("Original Shape :", X_scaled.shape)

print("Reduced Shape :", X_lda.shape)
```

```python
X_train_lda, X_test_lda, y_train, y_test = train_test_split(
    X_lda,
    y,
    test_size=0.2,
    random_state=42
)

model_lda = LogisticRegression()

model_lda.fit(X_train_lda, y_train)

pred_lda = model_lda.predict(X_test_lda)

print("Accuracy after LDA :")

print(accuracy_score(y_test, pred_lda))
```

```python
print("Classification Report (LDA)\n")

print(classification_report(y_test, pred_lda))
```
