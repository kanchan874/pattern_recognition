```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import (
    euclidean_distances,
    manhattan_distances,
    cosine_similarity
)
```

```python
df = pd.read_excel("Online Retail.xlsx")

print("Dataset Shape:", df.shape)
df.head()
```

```python
# Remove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Create TotalAmount column
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']

print(df.head())
```

```python
customer_profile = df.groupby('CustomerID').agg({
    'Quantity':'sum',
    'TotalAmount':'sum',
    'InvoiceNo':'nunique'
}).reset_index()

customer_profile.columns = [
    'CustomerID',
    'TotalQuantity',
    'TotalSpending',
    'TotalOrders'
]

customer_profile.head()
```

```python
features = customer_profile[
    ['TotalQuantity','TotalSpending','TotalOrders']
]

scaler = StandardScaler()

scaled_features = scaler.fit_transform(features)

print(scaled_features[:5])
```

```python
euclidean_matrix = euclidean_distances(scaled_features)

euclidean_df = pd.DataFrame(euclidean_matrix)

print("Euclidean Distance Matrix")
euclidean_df.head()
```

```python
manhattan_matrix = manhattan_distances(scaled_features)

manhattan_df = pd.DataFrame(manhattan_matrix)

print("Manhattan Distance Matrix")
manhattan_df.head()
```

```python
cosine_matrix = cosine_similarity(scaled_features)

cosine_df = pd.DataFrame(cosine_matrix)

print("Cosine Similarity Matrix")
cosine_df.head()
```

```python
customer_index = 0

similarity_scores = cosine_matrix[customer_index]

similar_customers = pd.DataFrame({
    'CustomerID': customer_profile['CustomerID'],
    'Similarity': similarity_scores
})

similar_customers = similar_customers.sort_values(
    by='Similarity',
    ascending=False
)

similar_customers.head(10)
```

```python
def edit_distance(str1, str2):

    m = len(str1)
    n = len(str2)

    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i

    for j in range(n+1):
        dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):

            if str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]

            else:
                dp[i][j] = 1 + min(
                    dp[i][j-1],
                    dp[i-1][j],
                    dp[i-1][j-1]
                )

    return dp[m][n]
```

```python
product1 = "WHITE HANGING HEART T-LIGHT HOLDER"
product2 = "WHITE METAL LANTERN"

distance = edit_distance(product1, product2)

print("Product 1:", product1)
print("Product 2:", product2)
print("Edit Distance =", distance)
```
